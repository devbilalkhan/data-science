### Find out the survivors based on age groups.

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

```python
df["age_groups"] = pd.cut(df["age"], bins=[0,18,25,50,75, 85], labels=["babies", "children", "young adults", "adults", "elderly"])
df.head()
```

```python
age_groups_survivors_subset = df.loc[df["survived"]==1, ["survived","age_groups"]]
age_groups_survivors_subset.groupby("age_groups")["survived"].count()
```
