## Question
How to combine 3 columns in 1 column? Make a separate data for everything based on sex( male, female and transgender).


```python
import pandas as pd
df = pd.read_csv("data.csv")
df.head()
```

```python
# Merge rural gender columns to a new column 'sex' with sum values of each gender
all_gender_rural_subset_sum = df.melt(value_vars=["MALE (RURAL)", "FEMALE (RURAL)", "TRANSGENDER (RURAL)"], var_name="SEX (RURAL)").groupby("SEX (RURAL)").agg({"value": "sum"})

# Rename the mean column label
all_gender_rural_subset_sum.rename(columns={"value": "GENDER SUM (RURAL) "}, inplace=True)
all_gender_rural_subset_sum
```

```python
# Merge rural gender columns to a new column 'sex' with mean values of each gender
all_gender_rural_subset_mean = df.melt(value_vars=["MALE (RURAL)", "FEMALE (RURAL)", "TRANSGENDER (RURAL)"], var_name="SEX (RURAL)").groupby("SEX (RURAL)").agg({"value": "mean"})

# Rename the mean column label
all_gender_rural_subset_mean.rename(columns={"value": "GENDER MEAN (RURAL) "}, inplace=True)
all_gender_rural_subset_mean
```

```python
# Merge urban gender columns to a new column 'sex' with sum values of each gender
all_gender_urban_subset_sum = df.melt(value_vars=["MALE (URBAN)", "FEMALE (URBAN)", "TRANSGENDER (URBAN)"], var_name="SEX (URBAN)").groupby("SEX (URBAN)").agg({"value": "sum"})

# Rename the mean column label
all_gender_urban_subset_sum.rename(columns={"value": "GENDER SUM (URBAN)"}, inplace=True)
all_gender_urban_subset_sum
```

```python
# Merge urban gender columns to a new column 'sex' with mean values of each gender
all_gender_urban_subset_mean = df.melt(value_vars=["MALE (URBAN)", "FEMALE (URBAN)", "TRANSGENDER (URBAN)"], var_name="SEX (URBAN)").groupby("SEX (URBAN)").agg({"value": "mean"})

# Rename the mean column label
all_gender_urban_subset_mean.rename(columns={"value": "GENDER MEAN (URBAN)"}, inplace=True)
all_gender_urban_subset_mean
```
